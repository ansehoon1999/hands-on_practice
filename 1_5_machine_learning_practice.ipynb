- 단변량 회귀는 각 구역에서 하나의 값을 예측하는 것이고 다변량 회귀는 각 구역에서 여러 개의 값을 예측하는 것

- 두 개의 클래스로 구분 할 수 있는 것

2. 성능 측정 (k겹 교차 검증)
- 훈련 세트를 k개의 fold로 나눈 다음에 해당 fold를 제외한 것으로 훈련을 시킨 모델을 사용하는 것

- 재현율: 양성인 것들 중에서 예측했을 때 실제로 양성인 것

- OvR: 10개의 분류를 할 수 있다면 가장 높은 확률을 골라내는 것
- OvO: 각각의 분류기를 비교한 다음에 클래스를 골라내는 것

5. 다중 레이블 분류
- 예를 들어 특정 사진에 여러 사람이 있다면 각각의 사람에 대해 이진 꼬리표를 출력하는 것

- 한 레이블이 다중 클래스가 될 수 있도록 일반화하는 것

  - 미니 배치 경사 하강법: 미니 배치라고 부르는 임의의 작은 샘플로 나누고 난 후에 각 샘플 단위로 나누고 그레이디언트를 계산하는 것

- 라쏘 회귀: 절댓값을 통해서 규제
- 엘라스틱 넷: 릿지와 회귀를 혼합하여
